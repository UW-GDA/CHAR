```python
import geopandas as gpd
import pandas as pd
import numpy as np
import rioxarray as rxr
import xarray as xr
import pystac_client
import planetary_computer as pc
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
from shapely.geometry import box
import seaborn as sns
from scipy import stats
import requests
import warnings
warnings.filterwarnings('ignore')

# =============================================================================
# ELEVATION ANALYSIS FUNCTIONS (from previous artifact)
# =============================================================================

def get_srtm_elevation(fire_geometry, buffer_meters=5000):
    """Download SRTM 30m elevation data via OpenTopography API"""
    print("Downloading SRTM 30m elevation data...")
    
    # Buffer the fire geometry for context
    fire_utm = fire_geometry.to_crs('EPSG:3857')  # Web Mercator for buffering
    buffered = fire_utm.buffer(buffer_meters)
    buffered_wgs84 = buffered.to_crs('EPSG:4326')
    
    # Get bounding box
    bounds = buffered_wgs84.bounds
    if hasattr(bounds, 'iloc'):
        bounds = bounds.iloc[0]  # If it's a series
    
    # Try OpenTopography first
    url = "https://cloud.sdsc.edu/v1/raster"
    
    params = {
        'raster': 'SRTM_GL1',  # SRTM 30m global
        'south': bounds[1],
        'north': bounds[3], 
        'west': bounds[0],
        'east': bounds[2],
        'output': 'GTiff',
        'API_Key': 'demoapikeyot2022'  # Demo key
    }
    
    try:
        print(f"Requesting elevation data for bounds: {bounds}")
        response = requests.get(url, params=params, timeout=120)
        
        if response.status_code == 200:
            # Save temporarily and load with rioxarray
            temp_file = 'temp_elevation.tif'
            with open(temp_file, 'wb') as f:
                f.write(response.content)
            
            elevation = rxr.open_rasterio(temp_file)
            elevation = elevation.squeeze()  # Remove band dimension
            
            # Clean up temp file
            import os
            os.remove(temp_file)
            
            print(f"✓ Downloaded elevation data: {elevation.shape} pixels")
            print(f"  Elevation range: {float(elevation.min().values):.0f}m to {float(elevation.max().values):.0f}m")
            
            return elevation
            
        else:
            print(f"❌ OpenTopography API failed: {response.status_code}")
            if response.status_code == 401:
                print("   API key may be invalid or rate limited")
            print("   Continuing with sample elevation data...")
            return None
            
    except Exception as e:
        print(f"❌ Error downloading SRTM data: {e}")
        print("   Continuing with sample elevation data...")
        return None

def analyze_fire_severity_elevation(dnbr, elevation, fire_geometry):
    """Analyze relationship between fire severity (dNBR) and elevation"""
    print("\nAnalyzing fire severity vs elevation relationship...")
    
    # Ensure both datasets have same CRS
    if dnbr.rio.crs != elevation.rio.crs:
        print("Reprojecting elevation to match dNBR...")
        elevation = elevation.rio.reproject_match(dnbr)
    
    # Clip both datasets to exact fire boundary to ensure same spatial extent
    fire_proj = gpd.GeoDataFrame([1], geometry=[fire_geometry], crs='EPSG:4326')
    fire_proj = fire_proj.to_crs(dnbr.rio.crs)
    
    print(f"Original shapes - dNBR: {dnbr.shape}, Elevation: {elevation.shape}")
    
    # Clip both datasets to the same fire boundary
    dnbr_fire = dnbr.rio.clip(fire_proj.geometry, drop=True, all_touched=True)
    elevation_fire = elevation.rio.clip(fire_proj.geometry, drop=True, all_touched=True)
    
    print(f"After clipping - dNBR: {dnbr_fire.shape}, Elevation: {elevation_fire.shape}")
    
    # If shapes still don't match, reproject elevation to exactly match dNBR grid
    if dnbr_fire.shape != elevation_fire.shape:
        print("Shapes don't match after clipping, reprojecting elevation to match dNBR grid...")
        elevation_fire = elevation_fire.rio.reproject_match(dnbr_fire)
        print(f"After reprojection - dNBR: {dnbr_fire.shape}, Elevation: {elevation_fire.shape}")
    
    # Flatten arrays and remove NaN values
    dnbr_values = dnbr_fire.values.flatten()
    elev_values = elevation_fire.values.flatten()
    
    print(f"Flattened arrays - dNBR: {dnbr_values.shape}, Elevation: {elev_values.shape}")
    
    # Create mask for valid data in both arrays
    valid_mask = ~(np.isnan(dnbr_values) | np.isnan(elev_values))
    dnbr_clean = dnbr_values[valid_mask]
    elev_clean = elev_values[valid_mask]
    
    if len(dnbr_clean) == 0:
        print("❌ No overlapping valid data between dNBR and elevation")
        return None
    
    if len(dnbr_clean) < 100:
        print(f"⚠️  Warning: Only {len(dnbr_clean)} valid pixels for analysis")
        print("   Results may not be statistically robust")
    
    print(f"✓ Analyzing {len(dnbr_clean)} valid pixels")
    
    # Calculate statistics
    try:
        correlation, p_value = stats.pearsonr(dnbr_clean, elev_clean)
    except Exception as e:
        print(f"❌ Error calculating correlation: {e}")
        return None
    
    # Elevation binning analysis
    n_bins = 8
    elev_bins = np.linspace(elev_clean.min(), elev_clean.max(), n_bins + 1)
    elev_centers = (elev_bins[:-1] + elev_bins[1:]) / 2
    
    severity_by_elevation = []
    
    for i in range(n_bins):
        mask = (elev_clean >= elev_bins[i]) & (elev_clean < elev_bins[i + 1])
        if i == n_bins - 1:  # Include upper bound in last bin
            mask = (elev_clean >= elev_bins[i]) & (elev_clean <= elev_bins[i + 1])
        
        if np.sum(mask) > 0:
            bin_dnbr = dnbr_clean[mask]
            
            # Calculate severity percentages for this elevation bin
            high_sev = np.sum(bin_dnbr > 0.66) / len(bin_dnbr) * 100
            mod_high_sev = np.sum((bin_dnbr >= 0.44) & (bin_dnbr <= 0.66)) / len(bin_dnbr) * 100
            mod_low_sev = np.sum((bin_dnbr >= 0.25) & (bin_dnbr < 0.44)) / len(bin_dnbr) * 100
            low_sev = np.sum((bin_dnbr >= 0.1) & (bin_dnbr < 0.25)) / len(bin_dnbr) * 100
            unburned = np.sum(bin_dnbr < 0.1) / len(bin_dnbr) * 100
            
            severity_by_elevation.append({
                'elevation': elev_centers[i],
                'high_severity': high_sev,
                'moderate_high': mod_high_sev,
                'moderate_low': mod_low_sev,
                'low_severity': low_sev,
                'unburned': unburned,
                'mean_dnbr': np.mean(bin_dnbr),
                'pixel_count': len(bin_dnbr)
            })
    
    results = {
        'dnbr_values': dnbr_clean,
        'elevation_values': elev_clean,
        'correlation': correlation,
        'p_value': p_value,
        'elevation_bins': elev_centers,
        'severity_by_elevation': pd.DataFrame(severity_by_elevation)
    }
    
    return results

def plot_elevation_analysis(results, fire_name="Fire"):
    """Create comprehensive plots of fire severity vs elevation"""
    fig = plt.figure(figsize=(16, 10))
    
    # 1. Scatter plot with trend line
    ax1 = plt.subplot(2, 3, 1)
    plt.scatter(results['elevation_values'], results['dnbr_values'], 
                alpha=0.5, s=2, c='red')
    
    # Add trend line
    z = np.polyfit(results['elevation_values'], results['dnbr_values'], 1)
    p = np.poly1d(z)
    x_trend = np.linspace(results['elevation_values'].min(), 
                         results['elevation_values'].max(), 100)
    plt.plot(x_trend, p(x_trend), 'blue', linewidth=2, alpha=0.8)
    
    plt.xlabel('Elevation (m)')
    plt.ylabel('dNBR')
    plt.title(f'Fire Severity vs Elevation\nr = {results["correlation"]:.3f}, p = {results["p_value"]:.4f}')
    plt.grid(True, alpha=0.3)
    
    # 2. Mean dNBR by elevation
    ax2 = plt.subplot(2, 3, 2)
    severity_df = results['severity_by_elevation']
    plt.plot(severity_df['elevation'], severity_df['mean_dnbr'], 
             'o-', linewidth=2, markersize=8, color='darkred')
    plt.xlabel('Elevation (m)')
    plt.ylabel('Mean dNBR')
    plt.title('Mean Fire Severity by Elevation')
    plt.grid(True, alpha=0.3)
    
    # 3. Stacked bar chart of severity classes
    ax3 = plt.subplot(2, 3, 3)
    width = 0.8
    x_pos = np.arange(len(severity_df))
    
    plt.bar(x_pos, severity_df['high_severity'], width, label='High Severity (>0.66)', color='darkred')
    plt.bar(x_pos, severity_df['moderate_high'], width, 
            bottom=severity_df['high_severity'], label='Moderate-High (0.44-0.66)', color='red')
    plt.bar(x_pos, severity_df['moderate_low'], width, 
            bottom=severity_df['high_severity'] + severity_df['moderate_high'], 
            label='Moderate-Low (0.25-0.44)', color='orange')
    plt.bar(x_pos, severity_df['low_severity'], width, 
            bottom=severity_df['high_severity'] + severity_df['moderate_high'] + severity_df['moderate_low'], 
            label='Low Severity (0.1-0.25)', color='yellow')
    plt.bar(x_pos, severity_df['unburned'], width,
            bottom=severity_df['high_severity'] + severity_df['moderate_high'] + 
                   severity_df['moderate_low'] + severity_df['low_severity'],
            label='Unburned (<0.1)', color='green')
    
    plt.xlabel('Elevation Bin')
    plt.ylabel('Percentage of Pixels')
    plt.title('Burn Severity Classes by Elevation')
    plt.xticks(x_pos, [f"{elev:.0f}m" for elev in severity_df['elevation']], rotation=45)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
    
    # 4. Elevation histogram
    ax4 = plt.subplot(2, 3, 4)
    plt.hist(results['elevation_values'], bins=30, alpha=0.7, color='brown', edgecolor='black')
    plt.xlabel('Elevation (m)')
    plt.ylabel('Number of Pixels')
    plt.title('Elevation Distribution in Fire Area')
    plt.grid(True, alpha=0.3)
    
    # 5. dNBR histogram
    ax5 = plt.subplot(2, 3, 5)
    plt.hist(results['dnbr_values'], bins=30, alpha=0.7, color='red', edgecolor='black')
    plt.xlabel('dNBR')
    plt.ylabel('Number of Pixels')
    plt.title('dNBR Distribution')
    plt.grid(True, alpha=0.3)
    
    # 6. Summary statistics
    ax6 = plt.subplot(2, 3, 6)
    ax6.axis('off')
    
    # Determine correlation strength and direction
    corr_abs = abs(results['correlation'])
    if corr_abs > 0.7:
        strength = "Strong"
    elif corr_abs > 0.3:
        strength = "Moderate"
    else:
        strength = "Weak"
    
    direction = "positive" if results['correlation'] > 0 else "negative"
    significance = "significant" if results['p_value'] < 0.05 else "not significant"
    
    summary_text = f"""ELEVATION ANALYSIS SUMMARY

Fire: {fire_name}
Total Pixels: {len(results['dnbr_values']):,}

Elevation Statistics:
• Min: {results['elevation_values'].min():.0f} m
• Max: {results['elevation_values'].max():.0f} m  
• Mean: {results['elevation_values'].mean():.0f} m
• Range: {results['elevation_values'].max() - results['elevation_values'].min():.0f} m

dNBR Statistics:
• Min: {results['dnbr_values'].min():.3f}
• Max: {results['dnbr_values'].max():.3f}
• Mean: {results['dnbr_values'].mean():.3f}

Correlation Analysis:
• Pearson r: {results['correlation']:.4f}
• P-value: {results['p_value']:.6f}
• Relationship: {strength} {direction}
• Statistical significance: {significance}

Interpretation:
{strength} {direction} correlation between 
elevation and fire severity
({significance} at α=0.05)"""
    
    ax6.text(0.05, 0.95, summary_text, transform=ax6.transAxes, fontsize=10,
             verticalalignment='top', fontfamily='monospace',
             bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.8))
    
    plt.suptitle(f'{fire_name} - Fire Severity vs Elevation Analysis', fontsize=16, y=0.98)
    plt.tight_layout()
    plt.subplots_adjust(top=0.93)
    plt.show()
    
    return fig

# =============================================================================
# COMPLETE EXAMPLE: FIRE SEVERITY + ELEVATION ANALYSIS
# =============================================================================

def complete_fire_analysis_example():
    """
    Complete example showing fire severity analysis with elevation
    This simulates your workflow with a real fire
    """
    
    print("="*80)
    print("COMPLETE FIRE SEVERITY + ELEVATION ANALYSIS EXAMPLE")
    print("="*80)
    
    # =============================================================================
    # 1. LOAD AND SELECT A FIRE (your existing code)
    # =============================================================================
    
    print("\n1. LOADING FIRE DATA...")
    try:
        fires = gpd.read_file("../downloaded_data/mnf_fires_all.geojson")
        
        # Clean fire year data (your existing logic)
        def clean_fire_year(year_value):
            try:
                if pd.isna(year_value):
                    return None
                year_str = str(year_value)
                if '-' in year_str:
                    return int(year_str.split('-')[0])
                return int(float(year_str))
            except (ValueError, TypeError):
                return None
        
        fires['FIRE_YEAR_CLEAN'] = fires['FIRE_YEAR'].apply(clean_fire_year)
        fires_clean = fires.dropna(subset=['FIRE_YEAR_CLEAN'])
        fires_clean['FIRE_YEAR_CLEAN'] = fires_clean['FIRE_YEAR_CLEAN'].astype(int)
        
        # Select 1989 fires (or nearby year if none found)
        fires_1989 = fires_clean[fires_clean['FIRE_YEAR_CLEAN'] == 1989].copy()
        
        if len(fires_1989) == 0:
            print("No 1989 fires found, trying nearby years...")
            for test_year in [1988, 1990, 1987, 1991]:
                test_fires = fires_clean[fires_clean['FIRE_YEAR_CLEAN'] == test_year]
                if len(test_fires) > 0:
                    print(f"Using {test_year} fires for analysis ({len(test_fires)} fires)")
                    fires_1989 = test_fires.copy()
                    break
        
        fires_1989_wgs84 = fires_1989.to_crs('EPSG:4326')
        
        # Select first fire for analysis
        test_fire = fires_1989_wgs84.iloc[0]
        fire_geom = test_fire.geometry
        fire_id = test_fire.get("UNQE_FIRE_", "unknown")
        incident = test_fire.get("INCIDENT", "Unknown Fire")
        fire_year = int(test_fire["FIRE_YEAR_CLEAN"])
        
        print(f"✓ Selected Fire: {incident} ({fire_id})")
        print(f"  Year: {fire_year}")
        print(f"  Area: {fire_geom.area:.6f} square degrees")
        
    except Exception as e:
        print(f"❌ Error loading fire data: {e}")
        print("Creating a sample fire geometry for demonstration...")
        
        # Create a sample fire polygon (you can replace with your actual fire)
        from shapely.geometry import Polygon
        
        # Sample coordinates - replace with your actual fire location
        sample_coords = [
            (-121.5, 47.5), (-121.45, 47.5), (-121.45, 47.55), 
            (-121.5, 47.55), (-121.5, 47.5)
        ]
        fire_geom = Polygon(sample_coords)
        incident = "Sample Fire"
        fire_id = "SAMPLE_001"
        fire_year = 1989
        
        print(f"✓ Using sample fire geometry for demonstration")
    
    # =============================================================================
    # 2. CREATE SAMPLE dNBR DATA (replace with your Landsat analysis)
    # =============================================================================
    
    print(f"\n2. CREATING SAMPLE dNBR DATA...")
    print("   (In real analysis, this would come from your Landsat NBR calculation)")
    
    # Create a realistic dNBR dataset for demonstration
    # In your real code, this would be the result of your Landsat analysis
    
    # Get fire bounds for creating sample data
    bounds = fire_geom.bounds
    
    # Create sample grid
    resolution = 0.0003  # ~30m at this latitude (approximate)
    x_coords = np.arange(bounds[0], bounds[2], resolution)
    y_coords = np.arange(bounds[1], bounds[3], resolution)
    
    # Create sample dNBR values with realistic patterns
    np.random.seed(42)  # For reproducible results
    
    dnbr_data = np.random.beta(2, 3, size=(len(y_coords), len(x_coords))) * 1.2  # Beta distribution for realistic dNBR
    dnbr_data = np.where(dnbr_data > 1.0, 1.0, dnbr_data)  # Cap at 1.0
    
    # Create xarray dataset
    dnbr = xr.DataArray(
        dnbr_data,
        coords={'y': y_coords, 'x': x_coords},
        dims=['y', 'x'],
        attrs={
            'long_name': 'Sample dNBR (Fire Boundary Only)',
            'description': 'Pre-fire NBR - Post-fire NBR (simulated)',
            'fire_id': fire_id,
            'incident_name': incident
        }
    )
    
    # Set CRS
    dnbr = dnbr.rio.write_crs('EPSG:4326')
    
    # Clip to fire boundary
    fire_gdf = gpd.GeoDataFrame([1], geometry=[fire_geom], crs='EPSG:4326')
    dnbr = dnbr.rio.clip(fire_gdf.geometry, drop=True)
    
    print(f"✓ Sample dNBR created: {dnbr.shape} pixels")
    print(f"  dNBR range: {float(dnbr.min().values):.3f} to {float(dnbr.max().values):.3f}")
    print(f"  Mean dNBR: {float(dnbr.mean().values):.3f}")
    
    # =============================================================================
    # 3. DOWNLOAD ELEVATION DATA
    # =============================================================================
    
    print(f"\n3. DOWNLOADING ELEVATION DATA...")
    fire_gdf = gpd.GeoDataFrame([1], geometry=[fire_geom], crs='EPSG:4326')
    elevation = get_srtm_elevation(fire_gdf, buffer_meters=3000)
    
    if elevation is None:
        print("❌ Could not download elevation data. Creating sample elevation for demonstration...")
        
        # Create sample elevation data that matches dNBR spatial grid exactly
        print(f"Creating sample elevation to match dNBR grid: {dnbr.shape}")
        
        # Get the same coordinates as dNBR
        y_coords = dnbr.y.values
        x_coords = dnbr.x.values
        
        # Create realistic elevation pattern
        Y, X = np.meshgrid(y_coords, x_coords, indexing='ij')
        
        # Create elevation pattern with some realistic topographic variation
        elev_data = (1000 + 
                    500 * np.sin(2 * np.pi * (Y - Y.min()) / (Y.max() - Y.min())) + 
                    300 * np.cos(2 * np.pi * (X - X.min()) / (X.max() - X.min())) + 
                    np.random.normal(0, 50, Y.shape))
        
        elevation = xr.DataArray(
            elev_data,
            coords={'y': y_coords, 'x': x_coords}, 
            dims=['y', 'x']
        ).rio.write_crs('EPSG:4326')
        
        print(f"✓ Sample elevation created: {elevation.shape} pixels")
        print(f"  Elevation range: {float(elevation.min().values):.0f}m to {float(elevation.max().values):.0f}m")
    
    # =============================================================================
    # 4. PERFORM ELEVATION ANALYSIS
    # =============================================================================
    
    print(f"\n4. PERFORMING ELEVATION ANALYSIS...")
    results = analyze_fire_severity_elevation(dnbr, elevation, fire_geom)
    
    if results is not None:
        # Create visualizations
        print(f"\n5. CREATING VISUALIZATIONS...")
        plot_elevation_analysis(results, f"{incident} ({fire_year})")
        
        # Print detailed results
        print(f"\n{'='*60}")
        print("ELEVATION ANALYSIS RESULTS")
        print(f"{'='*60}")
        
        print(f"\nFire: {incident} ({fire_id})")
        print(f"Year: {fire_year}")
        print(f"Total analyzed pixels: {len(results['dnbr_values']):,}")
        
        print(f"\nElevation Summary:")
        print(f"  Range: {results['elevation_values'].min():.0f} - {results['elevation_values'].max():.0f} m")
        print(f"  Mean: {results['elevation_values'].mean():.0f} m")
        print(f"  Std Dev: {results['elevation_values'].std():.0f} m")
        
        print(f"\nFire Severity Summary:")
        print(f"  dNBR Range: {results['dnbr_values'].min():.3f} - {results['dnbr_values'].max():.3f}")
        print(f"  Mean dNBR: {results['dnbr_values'].mean():.3f}")
        
        print(f"\nCorrelation Analysis:")
        print(f"  Pearson correlation: {results['correlation']:.4f}")
        print(f"  P-value: {results['p_value']:.6f}")
        print(f"  Statistical significance: {'Yes' if results['p_value'] < 0.05 else 'No'} (α=0.05)")
        
        # Interpretation
        corr_abs = abs(results['correlation'])
        if corr_abs > 0.3:
            strength = "strong" if corr_abs > 0.7 else "moderate"
        else:
            strength = "weak"
        
        direction = "positive" if results['correlation'] > 0 else "negative"
        
        print(f"\nInterpretation:")
        print(f"  There is a {strength} {direction} correlation between elevation and fire severity.")
        
        if results['correlation'] > 0.1:
            print(f"  Higher elevations tend to experience more severe burning.")
        elif results['correlation'] < -0.1:
            print(f"  Higher elevations tend to experience less severe burning.")
        else:
            print(f"  Elevation does not strongly influence fire severity in this fire.")
        
        # Elevation-based summary table
        severity_df = results['severity_by_elevation']
        print(f"\nSeverity by Elevation Zones:")
        print(f"{'Elevation (m)':<12} {'Mean dNBR':<10} {'High Sev %':<11} {'Mod-High %':<11} {'Pixels':<8}")
        print("-" * 65)
        for _, row in severity_df.iterrows():
            print(f"{row['elevation']:<12.0f} {row['mean_dnbr']:<10.3f} "
                  f"{row['high_severity']:<11.1f} {row['moderate_high']:<11.1f} {row['pixel_count']:<8.0f}")
        
        print(f"\n✓ ANALYSIS COMPLETE!")
        print(f"{'='*60}")
        
        return results
    
    else:
        print("❌ Elevation analysis failed")
        return None

# =============================================================================
# RUN THE COMPLETE EXAMPLE
# =============================================================================

if __name__ == "__main__":
    print("Running complete fire severity + elevation analysis example...")
    print("This demonstrates how to integrate elevation analysis with your existing fire severity code.")
    print("\nNote: This example uses sample data for demonstration.")
    print("In your real analysis, replace the sample dNBR creation with your actual Landsat NBR calculation.\n")
    
    # Run the complete analysis
    results = complete_fire_analysis_example()
    
    print("\n" + "="*80)
    print("EXAMPLE COMPLETE!")
    print("="*80)
    print("\nTo use with your real data:")
    print("1. Replace the sample dNBR creation with your actual Landsat analysis")
    print("2. Use your real fire geometry from the geojson file")
    print("3. The elevation analysis will work automatically with real SRTM data")
    print("\nKey functions you can use:")
    print("- get_srtm_elevation(): Downloads elevation data")
    print("- analyze_fire_severity_elevation(): Performs statistical analysis")
    print("- plot_elevation_analysis(): Creates comprehensive visualizations")
```

```python
import geopandas as gpd
import pandas as pd
import numpy as np
import rioxarray as rxr
import xarray as xr
import pystac_client
import planetary_computer as pc
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
from shapely.geometry import box
import seaborn as sns
from scipy import stats
import requests
import warnings
warnings.filterwarnings('ignore')

# =============================================================================
# ELEVATION ANALYSIS FUNCTIONS (from previous artifact)
# =============================================================================

def get_srtm_elevation_alternative(fire_geometry, buffer_meters=5000):
    """
    Alternative elevation data sources if OpenTopography fails
    """
    print("Trying alternative elevation data sources...")
    
    # Get bounds
    fire_utm = fire_geometry.to_crs('EPSG:3857')
    buffered = fire_utm.buffer(buffer_meters)
    buffered_wgs84 = buffered.to_crs('EPSG:4326')
    bounds = buffered_wgs84.bounds
    if hasattr(bounds, 'iloc'):
        bounds = bounds.iloc[0]
    
    # Option 1: Try OpenTopoData API (different service)
    try:
        print("Trying OpenTopoData API...")
        
        # Create a small grid of points for elevation sampling
        lat_points = np.linspace(bounds[1], bounds[3], 20)
        lon_points = np.linspace(bounds[0], bounds[2], 20)
        
        elevations = []
        locations = []
        
        # Sample elevation at grid points (this API works with point queries)
        for lat in lat_points[::2]:  # Sample every other point to reduce requests
            for lon in lon_points[::2]:
                url = f"https://api.opentopodata.org/v1/srtm30m?locations={lat},{lon}"
                try:
                    response = requests.get(url, timeout=10)
                    if response.status_code == 200:
                        data = response.json()
                        if data['status'] == 'OK' and len(data['results']) > 0:
                            elev = data['results'][0]['elevation']
                            if elev is not None:
                                elevations.append(elev)
                                locations.append((lon, lat))
                except:
                    continue
                
                # Small delay to respect rate limits
                import time
                time.sleep(0.1)
        
        if len(elevations) > 10:
            print(f"✓ Retrieved {len(elevations)} elevation points from OpenTopoData")
            
            # Interpolate to create a grid
            from scipy.interpolate import griddata
            
            # Create target grid matching fire area
            target_lats = np.linspace(bounds[1], bounds[3], 100)
            target_lons = np.linspace(bounds[0], bounds[2], 100)
            target_lon_grid, target_lat_grid = np.meshgrid(target_lons, target_lats)
            
            # Interpolate elevations
            points = np.array(locations)
            values = np.array(elevations)
            
            interpolated = griddata(points, values, (target_lon_grid, target_lat_grid), method='cubic')
            
            # Create xarray dataset
            elevation = xr.DataArray(
                interpolated,
                coords={'y': target_lats, 'x': target_lons},
                dims=['y', 'x']
            ).rio.write_crs('EPSG:4326')
            
            print(f"✓ Created interpolated elevation grid: {elevation.shape}")
            return elevation
            
    except Exception as e:
        print(f"OpenTopoData failed: {e}")
    
    # Option 2: Try a different OpenTopography endpoint
    try:
        print("Trying OpenTopography without API key...")
        url = "https://cloud.sdsc.edu/v1/raster"
        
        params = {
            'raster': 'SRTM_GL1',
            'south': bounds[1],
            'north': bounds[3], 
            'west': bounds[0],
            'east': bounds[2],
            'output': 'GTiff'
            # No API key - try without it
        }
        
        response = requests.get(url, params=params, timeout=120)
        
        if response.status_code == 200:
            temp_file = 'temp_elevation_alt.tif'
            with open(temp_file, 'wb') as f:
                f.write(response.content)
            
            elevation = rxr.open_rasterio(temp_file)
            elevation = elevation.squeeze()
            
            import os
            os.remove(temp_file)
            
            print(f"✓ OpenTopography without API key succeeded: {elevation.shape}")
            return elevation
    except Exception as e:
        print(f"OpenTopography without key failed: {e}")
    
    # Option 3: Instructions for manual download
    print("\n" + "="*60)
    print("MANUAL ELEVATION DATA OPTIONS:")
    print("="*60)
    print("1. Get free OpenTopography API key:")
    print("   - Visit: https://portal.opentopography.org/")
    print("   - Register (free)")
    print("   - Replace 'demoapikeyot2022' with your key")
    
    print("\n2. Download SRTM manually:")
    print("   - Visit: https://earthexplorer.usgs.gov/")
    print("   - Search for SRTM data in your fire area")
    print("   - Use load_local_dem() function")
    
    print(f"\n3. Your fire bounds for manual download:")
    print(f"   North: {bounds[3]:.6f}")
    print(f"   South: {bounds[1]:.6f}")
    print(f"   East:  {bounds[2]:.6f}")
    print(f"   West:  {bounds[0]:.6f}")
    print("="*60)
    
    return None
    """Download SRTM 30m elevation data via OpenTopography API"""
    print("Downloading SRTM 30m elevation data...")
    
    # Buffer the fire geometry for context
    fire_utm = fire_geometry.to_crs('EPSG:3857')  # Web Mercator for buffering
    buffered = fire_utm.buffer(buffer_meters)
    buffered_wgs84 = buffered.to_crs('EPSG:4326')
    
    # Get bounding box
    bounds = buffered_wgs84.bounds
    if hasattr(bounds, 'iloc'):
        bounds = bounds.iloc[0]  # If it's a series
    
    # Try OpenTopography first
    url = "https://cloud.sdsc.edu/v1/raster"
    
    params = {
        'raster': 'SRTM_GL1',  # SRTM 30m global
        'south': bounds[1],
        'north': bounds[3], 
        'west': bounds[0],
        'east': bounds[2],
        'output': 'GTiff',
        'API_Key': 'demoapikeyot2022'  # Demo key
    }
    
    try:
        print(f"Requesting elevation data for bounds: {bounds}")
        response = requests.get(url, params=params, timeout=120)
        
        if response.status_code == 200:
            # Save temporarily and load with rioxarray
            temp_file = 'temp_elevation.tif'
            with open(temp_file, 'wb') as f:
                f.write(response.content)
            
            elevation = rxr.open_rasterio(temp_file)
            elevation = elevation.squeeze()  # Remove band dimension
            
            # Clean up temp file
            import os
            os.remove(temp_file)
            
            print(f"✓ Downloaded elevation data: {elevation.shape} pixels")
            print(f"  Elevation range: {float(elevation.min().values):.0f}m to {float(elevation.max().values):.0f}m")
            
            return elevation
            
        else:
            print(f"❌ OpenTopography API failed: {response.status_code}")
            if response.status_code == 401:
                print("   API key may be invalid or rate limited")
            print("   Continuing with sample elevation data...")
            return None
            
    except Exception as e:
        print(f"❌ Error downloading SRTM data: {e}")
        print("   Continuing with sample elevation data...")
        return None

def analyze_fire_severity_elevation(dnbr, elevation, fire_geometry):
    """Analyze relationship between fire severity (dNBR) and elevation"""
    print("\nAnalyzing fire severity vs elevation relationship...")
    
    # Ensure both datasets have same CRS
    if dnbr.rio.crs != elevation.rio.crs:
        print("Reprojecting elevation to match dNBR...")
        elevation = elevation.rio.reproject_match(dnbr)
    
    # Clip both datasets to exact fire boundary to ensure same spatial extent
    fire_proj = gpd.GeoDataFrame([1], geometry=[fire_geometry], crs='EPSG:4326')
    fire_proj = fire_proj.to_crs(dnbr.rio.crs)
    
    print(f"Original shapes - dNBR: {dnbr.shape}, Elevation: {elevation.shape}")
    
    # Clip both datasets to the same fire boundary
    dnbr_fire = dnbr.rio.clip(fire_proj.geometry, drop=True, all_touched=True)
    elevation_fire = elevation.rio.clip(fire_proj.geometry, drop=True, all_touched=True)
    
    print(f"After clipping - dNBR: {dnbr_fire.shape}, Elevation: {elevation_fire.shape}")
    
    # If shapes still don't match, reproject elevation to exactly match dNBR grid
    if dnbr_fire.shape != elevation_fire.shape:
        print("Shapes don't match after clipping, reprojecting elevation to match dNBR grid...")
        elevation_fire = elevation_fire.rio.reproject_match(dnbr_fire)
        print(f"After reprojection - dNBR: {dnbr_fire.shape}, Elevation: {elevation_fire.shape}")
    
    # Flatten arrays and remove NaN values
    dnbr_values = dnbr_fire.values.flatten()
    elev_values = elevation_fire.values.flatten()
    
    print(f"Flattened arrays - dNBR: {dnbr_values.shape}, Elevation: {elev_values.shape}")
    
    # Create mask for valid data in both arrays
    valid_mask = ~(np.isnan(dnbr_values) | np.isnan(elev_values))
    dnbr_clean = dnbr_values[valid_mask]
    elev_clean = elev_values[valid_mask]
    
    if len(dnbr_clean) == 0:
        print("❌ No overlapping valid data between dNBR and elevation")
        return None
    
    if len(dnbr_clean) < 100:
        print(f"⚠️  Warning: Only {len(dnbr_clean)} valid pixels for analysis")
        print("   Results may not be statistically robust")
    
    print(f"✓ Analyzing {len(dnbr_clean)} valid pixels")
    
    # Calculate statistics
    try:
        correlation, p_value = stats.pearsonr(dnbr_clean, elev_clean)
    except Exception as e:
        print(f"❌ Error calculating correlation: {e}")
        return None
    
    # Elevation binning analysis
    n_bins = 8
    elev_bins = np.linspace(elev_clean.min(), elev_clean.max(), n_bins + 1)
    elev_centers = (elev_bins[:-1] + elev_bins[1:]) / 2
    
    severity_by_elevation = []
    
    for i in range(n_bins):
        mask = (elev_clean >= elev_bins[i]) & (elev_clean < elev_bins[i + 1])
        if i == n_bins - 1:  # Include upper bound in last bin
            mask = (elev_clean >= elev_bins[i]) & (elev_clean <= elev_bins[i + 1])
        
        if np.sum(mask) > 0:
            bin_dnbr = dnbr_clean[mask]
            
            # Calculate severity percentages for this elevation bin
            high_sev = np.sum(bin_dnbr > 0.66) / len(bin_dnbr) * 100
            mod_high_sev = np.sum((bin_dnbr >= 0.44) & (bin_dnbr <= 0.66)) / len(bin_dnbr) * 100
            mod_low_sev = np.sum((bin_dnbr >= 0.25) & (bin_dnbr < 0.44)) / len(bin_dnbr) * 100
            low_sev = np.sum((bin_dnbr >= 0.1) & (bin_dnbr < 0.25)) / len(bin_dnbr) * 100
            unburned = np.sum(bin_dnbr < 0.1) / len(bin_dnbr) * 100
            
            severity_by_elevation.append({
                'elevation': elev_centers[i],
                'high_severity': high_sev,
                'moderate_high': mod_high_sev,
                'moderate_low': mod_low_sev,
                'low_severity': low_sev,
                'unburned': unburned,
                'mean_dnbr': np.mean(bin_dnbr),
                'pixel_count': len(bin_dnbr)
            })
    
    results = {
        'dnbr_values': dnbr_clean,
        'elevation_values': elev_clean,
        'correlation': correlation,
        'p_value': p_value,
        'elevation_bins': elev_centers,
        'severity_by_elevation': pd.DataFrame(severity_by_elevation)
    }
    
    return results

def plot_elevation_analysis(results, fire_name="Fire"):
    """Create comprehensive plots of fire severity vs elevation"""
    fig = plt.figure(figsize=(16, 10))
    
    # 1. Scatter plot with trend line
    ax1 = plt.subplot(2, 3, 1)
    plt.scatter(results['elevation_values'], results['dnbr_values'], 
                alpha=0.5, s=2, c='red')
    
    # Add trend line
    z = np.polyfit(results['elevation_values'], results['dnbr_values'], 1)
    p = np.poly1d(z)
    x_trend = np.linspace(results['elevation_values'].min(), 
                         results['elevation_values'].max(), 100)
    plt.plot(x_trend, p(x_trend), 'blue', linewidth=2, alpha=0.8)
    
    plt.xlabel('Elevation (m)')
    plt.ylabel('dNBR')
    plt.title(f'Fire Severity vs Elevation\nr = {results["correlation"]:.3f}, p = {results["p_value"]:.4f}')
    plt.grid(True, alpha=0.3)
    
    # 2. Mean dNBR by elevation
    ax2 = plt.subplot(2, 3, 2)
    severity_df = results['severity_by_elevation']
    plt.plot(severity_df['elevation'], severity_df['mean_dnbr'], 
             'o-', linewidth=2, markersize=8, color='darkred')
    plt.xlabel('Elevation (m)')
    plt.ylabel('Mean dNBR')
    plt.title('Mean Fire Severity by Elevation')
    plt.grid(True, alpha=0.3)
    
    # 3. Stacked bar chart of severity classes
    ax3 = plt.subplot(2, 3, 3)
    width = 0.8
    x_pos = np.arange(len(severity_df))
    
    plt.bar(x_pos, severity_df['high_severity'], width, label='High Severity (>0.66)', color='darkred')
    plt.bar(x_pos, severity_df['moderate_high'], width, 
            bottom=severity_df['high_severity'], label='Moderate-High (0.44-0.66)', color='red')
    plt.bar(x_pos, severity_df['moderate_low'], width, 
            bottom=severity_df['high_severity'] + severity_df['moderate_high'], 
            label='Moderate-Low (0.25-0.44)', color='orange')
    plt.bar(x_pos, severity_df['low_severity'], width, 
            bottom=severity_df['high_severity'] + severity_df['moderate_high'] + severity_df['moderate_low'], 
            label='Low Severity (0.1-0.25)', color='yellow')
    plt.bar(x_pos, severity_df['unburned'], width,
            bottom=severity_df['high_severity'] + severity_df['moderate_high'] + 
                   severity_df['moderate_low'] + severity_df['low_severity'],
            label='Unburned (<0.1)', color='green')
    
    plt.xlabel('Elevation Bin')
    plt.ylabel('Percentage of Pixels')
    plt.title('Burn Severity Classes by Elevation')
    plt.xticks(x_pos, [f"{elev:.0f}m" for elev in severity_df['elevation']], rotation=45)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
    
    # 4. Elevation histogram
    ax4 = plt.subplot(2, 3, 4)
    plt.hist(results['elevation_values'], bins=30, alpha=0.7, color='brown', edgecolor='black')
    plt.xlabel('Elevation (m)')
    plt.ylabel('Number of Pixels')
    plt.title('Elevation Distribution in Fire Area')
    plt.grid(True, alpha=0.3)
    
    # 5. dNBR histogram
    ax5 = plt.subplot(2, 3, 5)
    plt.hist(results['dnbr_values'], bins=30, alpha=0.7, color='red', edgecolor='black')
    plt.xlabel('dNBR')
    plt.ylabel('Number of Pixels')
    plt.title('dNBR Distribution')
    plt.grid(True, alpha=0.3)
    
    # 6. Summary statistics
    ax6 = plt.subplot(2, 3, 6)
    ax6.axis('off')
    
    # Determine correlation strength and direction
    corr_abs = abs(results['correlation'])
    if corr_abs > 0.7:
        strength = "Strong"
    elif corr_abs > 0.3:
        strength = "Moderate"
    else:
        strength = "Weak"
    
    direction = "positive" if results['correlation'] > 0 else "negative"
    significance = "significant" if results['p_value'] < 0.05 else "not significant"
    
    summary_text = f"""ELEVATION ANALYSIS SUMMARY

Fire: {fire_name}
Total Pixels: {len(results['dnbr_values']):,}

Elevation Statistics:
• Min: {results['elevation_values'].min():.0f} m
• Max: {results['elevation_values'].max():.0f} m  
• Mean: {results['elevation_values'].mean():.0f} m
• Range: {results['elevation_values'].max() - results['elevation_values'].min():.0f} m

dNBR Statistics:
• Min: {results['dnbr_values'].min():.3f}
• Max: {results['dnbr_values'].max():.3f}
• Mean: {results['dnbr_values'].mean():.3f}

Correlation Analysis:
• Pearson r: {results['correlation']:.4f}
• P-value: {results['p_value']:.6f}
• Relationship: {strength} {direction}
• Statistical significance: {significance}

Interpretation:
{strength} {direction} correlation between 
elevation and fire severity
({significance} at α=0.05)"""
    
    ax6.text(0.05, 0.95, summary_text, transform=ax6.transAxes, fontsize=10,
             verticalalignment='top', fontfamily='monospace',
             bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.8))
    
    plt.suptitle(f'{fire_name} - Fire Severity vs Elevation Analysis', fontsize=16, y=0.98)
    plt.tight_layout()
    plt.subplots_adjust(top=0.93)
    plt.show()
    
    return fig

# =============================================================================
# COMPLETE EXAMPLE: FIRE SEVERITY + ELEVATION ANALYSIS
# =============================================================================

def complete_fire_analysis_example():
    """
    Complete example showing fire severity analysis with elevation
    This simulates your workflow with a real fire
    """
    
    print("="*80)
    print("COMPLETE FIRE SEVERITY + ELEVATION ANALYSIS EXAMPLE")
    print("="*80)
    
    # =============================================================================
    # 1. LOAD AND SELECT A FIRE (your existing code)
    # =============================================================================
    
    print("\n1. LOADING FIRE DATA...")
    try:
        fires = gpd.read_file("../downloaded_data/mnf_fires_all.geojson")
        
        # Clean fire year data (your existing logic)
        def clean_fire_year(year_value):
            try:
                if pd.isna(year_value):
                    return None
                year_str = str(year_value)
                if '-' in year_str:
                    return int(year_str.split('-')[0])
                return int(float(year_str))
            except (ValueError, TypeError):
                return None
        
        fires['FIRE_YEAR_CLEAN'] = fires['FIRE_YEAR'].apply(clean_fire_year)
        fires_clean = fires.dropna(subset=['FIRE_YEAR_CLEAN'])
        fires_clean['FIRE_YEAR_CLEAN'] = fires_clean['FIRE_YEAR_CLEAN'].astype(int)
        
        # Select 1989 fires (or nearby year if none found)
        fires_1989 = fires_clean[fires_clean['FIRE_YEAR_CLEAN'] == 1989].copy()
        
        if len(fires_1989) == 0:
            print("No 1989 fires found, trying nearby years...")
            for test_year in [1988, 1990, 1987, 1991]:
                test_fires = fires_clean[fires_clean['FIRE_YEAR_CLEAN'] == test_year]
                if len(test_fires) > 0:
                    print(f"Using {test_year} fires for analysis ({len(test_fires)} fires)")
                    fires_1989 = test_fires.copy()
                    break
        
        fires_1989_wgs84 = fires_1989.to_crs('EPSG:4326')
        
        # Select first fire for analysis
        test_fire = fires_1989_wgs84.iloc[0]
        fire_geom = test_fire.geometry
        fire_id = test_fire.get("UNQE_FIRE_", "unknown")
        incident = test_fire.get("INCIDENT", "Unknown Fire")
        fire_year = int(test_fire["FIRE_YEAR_CLEAN"])
        
        print(f"✓ Selected Fire: {incident} ({fire_id})")
        print(f"  Year: {fire_year}")
        print(f"  Area: {fire_geom.area:.6f} square degrees")
        
    except Exception as e:
        print(f"❌ Error loading fire data: {e}")
        print("Creating a sample fire geometry for demonstration...")
        
        # Create a sample fire polygon (you can replace with your actual fire)
        from shapely.geometry import Polygon
        
        # Sample coordinates - replace with your actual fire location
        sample_coords = [
            (-121.5, 47.5), (-121.45, 47.5), (-121.45, 47.55), 
            (-121.5, 47.55), (-121.5, 47.5)
        ]
        fire_geom = Polygon(sample_coords)
        incident = "Sample Fire"
        fire_id = "SAMPLE_001"
        fire_year = 1989
        
        print(f"✓ Using sample fire geometry for demonstration")
    
    # =============================================================================
    # 2. CREATE SAMPLE dNBR DATA (replace with your Landsat analysis)
    # =============================================================================
    
    print(f"\n2. CREATING SAMPLE dNBR DATA...")
    print("   (In real analysis, this would come from your Landsat NBR calculation)")
    
    # Create a realistic dNBR dataset for demonstration
    # In your real code, this would be the result of your Landsat analysis
    
    # Get fire bounds for creating sample data
    bounds = fire_geom.bounds
    
    # Create sample grid
    resolution = 0.0003  # ~30m at this latitude (approximate)
    x_coords = np.arange(bounds[0], bounds[2], resolution)
    y_coords = np.arange(bounds[1], bounds[3], resolution)
    
    # Create sample dNBR values with realistic patterns
    np.random.seed(42)  # For reproducible results
    
    dnbr_data = np.random.beta(2, 3, size=(len(y_coords), len(x_coords))) * 1.2  # Beta distribution for realistic dNBR
    dnbr_data = np.where(dnbr_data > 1.0, 1.0, dnbr_data)  # Cap at 1.0
    
    # Create xarray dataset
    dnbr = xr.DataArray(
        dnbr_data,
        coords={'y': y_coords, 'x': x_coords},
        dims=['y', 'x'],
        attrs={
            'long_name': 'Sample dNBR (Fire Boundary Only)',
            'description': 'Pre-fire NBR - Post-fire NBR (simulated)',
            'fire_id': fire_id,
            'incident_name': incident
        }
    )
    
    # Set CRS
    dnbr = dnbr.rio.write_crs('EPSG:4326')
    
    # Clip to fire boundary
    fire_gdf = gpd.GeoDataFrame([1], geometry=[fire_geom], crs='EPSG:4326')
    dnbr = dnbr.rio.clip(fire_gdf.geometry, drop=True)
    
    print(f"✓ Sample dNBR created: {dnbr.shape} pixels")
    print(f"  dNBR range: {float(dnbr.min().values):.3f} to {float(dnbr.max().values):.3f}")
    print(f"  Mean dNBR: {float(dnbr.mean().values):.3f}")
    
    # =============================================================================
    # 3. DOWNLOAD ELEVATION DATA WITH ALTERNATIVES
    # =============================================================================
    
    print(f"\n3. DOWNLOADING ELEVATION DATA...")
    fire_gdf = gpd.GeoDataFrame([1], geometry=[fire_geom], crs='EPSG:4326')
    
    # Try primary method first
    elevation = get_srtm_elevation(fire_gdf, buffer_meters=3000)
    
    # If that fails, try alternatives
    if elevation is None:
        elevation = get_srtm_elevation_alternative(fire_gdf, buffer_meters=3000)
    
    if elevation is None:
        print("❌ Could not download elevation data. Creating sample elevation for demonstration...")
        
        # Create sample elevation data that matches dNBR spatial grid exactly
        print(f"Creating sample elevation to match dNBR grid: {dnbr.shape}")
        
        # Get the same coordinates as dNBR
        y_coords = dnbr.y.values
        x_coords = dnbr.x.values
        
        # Create realistic elevation pattern
        Y, X = np.meshgrid(y_coords, x_coords, indexing='ij')
        
        # Create elevation pattern with some realistic topographic variation
        elev_data = (1000 + 
                    500 * np.sin(2 * np.pi * (Y - Y.min()) / (Y.max() - Y.min())) + 
                    300 * np.cos(2 * np.pi * (X - X.min()) / (X.max() - X.min())) + 
                    np.random.normal(0, 50, Y.shape))
        
        elevation = xr.DataArray(
            elev_data,
            coords={'y': y_coords, 'x': x_coords}, 
            dims=['y', 'x']
        ).rio.write_crs('EPSG:4326')
        
        print(f"✓ Sample elevation created: {elevation.shape} pixels")
        print(f"  Elevation range: {float(elevation.min().values):.0f}m to {float(elevation.max().values):.0f}m")
    
    # =============================================================================
    # 4. PERFORM ELEVATION ANALYSIS
    # =============================================================================
    
    print(f"\n4. PERFORMING ELEVATION ANALYSIS...")
    results = analyze_fire_severity_elevation(dnbr, elevation, fire_geom)
    
    if results is not None:
        # Create visualizations
        print(f"\n5. CREATING VISUALIZATIONS...")
        plot_elevation_analysis(results, f"{incident} ({fire_year})")
        
        # Print detailed results
        print(f"\n{'='*60}")
        print("ELEVATION ANALYSIS RESULTS")
        print(f"{'='*60}")
        
        print(f"\nFire: {incident} ({fire_id})")
        print(f"Year: {fire_year}")
        print(f"Total analyzed pixels: {len(results['dnbr_values']):,}")
        
        print(f"\nElevation Summary:")
        print(f"  Range: {results['elevation_values'].min():.0f} - {results['elevation_values'].max():.0f} m")
        print(f"  Mean: {results['elevation_values'].mean():.0f} m")
        print(f"  Std Dev: {results['elevation_values'].std():.0f} m")
        
        print(f"\nFire Severity Summary:")
        print(f"  dNBR Range: {results['dnbr_values'].min():.3f} - {results['dnbr_values'].max():.3f}")
        print(f"  Mean dNBR: {results['dnbr_values'].mean():.3f}")
        
        print(f"\nCorrelation Analysis:")
        print(f"  Pearson correlation: {results['correlation']:.4f}")
        print(f"  P-value: {results['p_value']:.6f}")
        print(f"  Statistical significance: {'Yes' if results['p_value'] < 0.05 else 'No'} (α=0.05)")
        
        # Interpretation
        corr_abs = abs(results['correlation'])
        if corr_abs > 0.3:
            strength = "strong" if corr_abs > 0.7 else "moderate"
        else:
            strength = "weak"
        
        direction = "positive" if results['correlation'] > 0 else "negative"
        
        print(f"\nInterpretation:")
        print(f"  There is a {strength} {direction} correlation between elevation and fire severity.")
        
        if results['correlation'] > 0.1:
            print(f"  Higher elevations tend to experience more severe burning.")
        elif results['correlation'] < -0.1:
            print(f"  Higher elevations tend to experience less severe burning.")
        else:
            print(f"  Elevation does not strongly influence fire severity in this fire.")
        
        # Elevation-based summary table
        severity_df = results['severity_by_elevation']
        print(f"\nSeverity by Elevation Zones:")
        print(f"{'Elevation (m)':<12} {'Mean dNBR':<10} {'High Sev %':<11} {'Mod-High %':<11} {'Pixels':<8}")
        print("-" * 65)
        for _, row in severity_df.iterrows():
            print(f"{row['elevation']:<12.0f} {row['mean_dnbr']:<10.3f} "
                  f"{row['high_severity']:<11.1f} {row['moderate_high']:<11.1f} {row['pixel_count']:<8.0f}")
        
        print(f"\n✓ ANALYSIS COMPLETE!")
        print(f"{'='*60}")
        
        return results
    
    else:
        print("❌ Elevation analysis failed")
        return None

# =============================================================================
# RUN THE COMPLETE EXAMPLE
# =============================================================================

if __name__ == "__main__":
    print("Running complete fire severity + elevation analysis example...")
    print("This demonstrates how to integrate elevation analysis with your existing fire severity code.")
    print("\nNote: This example uses sample data for demonstration.")
    print("In your real analysis, replace the sample dNBR creation with your actual Landsat NBR calculation.\n")
    
    # Run the complete analysis
    results = complete_fire_analysis_example()
    
    print("\n" + "="*80)
    print("EXAMPLE COMPLETE!")
    print("="*80)
    print("\nTo use with your real data:")
    print("1. Replace the sample dNBR creation with your actual Landsat analysis")
    print("2. Use your real fire geometry from the geojson file")
    print("3. The elevation analysis will work automatically with real SRTM data")
    print("\nKey functions you can use:")
    print("- get_srtm_elevation(): Downloads elevation data")
    print("- analyze_fire_severity_elevation(): Performs statistical analysis")
    print("- plot_elevation_analysis(): Creates comprehensive visualizations")
```

```python
import geopandas as gpd
import pandas as pd
import numpy as np
import rioxarray as rxr
import xarray as xr
import pystac_client
import planetary_computer as pc
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
from shapely.geometry import box
import seaborn as sns
from scipy import stats
import requests
import warnings
warnings.filterwarnings('ignore')

# =============================================================================
# REAL FIRE DATA LOADING AND SELECTION
# =============================================================================

def load_and_select_fire(geojson_path, target_year=1989, fire_index=0):
    """
    Load fire data and select a specific fire for analysis
    """
    print("="*80)
    print("LOADING REAL FIRE DATA")
    print("="*80)
    
    # Load fire dataset
    print(f"Loading fire data from: {geojson_path}")
    fires = gpd.read_file(geojson_path)
    
    # Clean fire year data
    def clean_fire_year(year_value):
        try:
            if pd.isna(year_value):
                return None
            year_str = str(year_value)
            if '-' in year_str:
                return int(year_str.split('-')[0])
            return int(float(year_str))
        except (ValueError, TypeError):
            return None
    
    fires['FIRE_YEAR_CLEAN'] = fires['FIRE_YEAR'].apply(clean_fire_year)
    fires_clean = fires.dropna(subset=['FIRE_YEAR_CLEAN'])
    fires_clean['FIRE_YEAR_CLEAN'] = fires_clean['FIRE_YEAR_CLEAN'].astype(int)
    
    print(f"Available years: {sorted(fires_clean['FIRE_YEAR_CLEAN'].unique())}")
    
    # Filter for target year
    target_fires = fires_clean[fires_clean['FIRE_YEAR_CLEAN'] == target_year].copy()
    
    if len(target_fires) == 0:
        print(f"No fires found for {target_year}, trying nearby years...")
        for test_year in [target_year-1, target_year+1, target_year-2, target_year+2]:
            test_fires = fires_clean[fires_clean['FIRE_YEAR_CLEAN'] == test_year]
            if len(test_fires) > 0:
                print(f"Using {test_year} fires ({len(test_fires)} available)")
                target_fires = test_fires.copy()
                target_year = test_year
                break
    
    if len(target_fires) == 0:
        raise ValueError(f"No fires found near year {target_year}")
    
    # Convert to WGS84
    target_fires_wgs84 = target_fires.to_crs('EPSG:4326')
    
    # Show available fires
    print(f"\nAvailable fires for {target_year}:")
    for i, (idx, fire) in enumerate(target_fires_wgs84.iterrows()):
        fire_name = fire.get('INCIDENT', f'Fire_{i}')
        fire_id = fire.get('UNQE_FIRE_', f'ID_{i}')
        area = fire.geometry.area
        print(f"  {i}: {fire_name} ({fire_id}) - Area: {area:.8f} sq deg")
    
    # Select fire
    if fire_index >= len(target_fires_wgs84):
        fire_index = 0
        print(f"Requested index {fire_index} not available, using fire 0")
    
    selected_fire = target_fires_wgs84.iloc[fire_index]
    
    fire_info = {
        'geometry': selected_fire.geometry,
        'fire_id': selected_fire.get("UNQE_FIRE_", "unknown"),
        'incident_name': selected_fire.get("INCIDENT", "Unknown Fire"),
        'fire_year': int(selected_fire["FIRE_YEAR_CLEAN"]),
        'index': fire_index
    }
    
    print(f"\n✓ SELECTED FIRE:")
    print(f"  Name: {fire_info['incident_name']}")
    print(f"  ID: {fire_info['fire_id']}")
    print(f"  Year: {fire_info['fire_year']}")
    print(f"  Area: {fire_info['geometry'].area:.8f} square degrees")
    
    return fire_info

# =============================================================================
# LANDSAT DATA SEARCH AND NBR CALCULATION
# =============================================================================

def search_landsat_imagery(geometry, target_date, days_buffer=60, platform="landsat-5"):
    """Search for Landsat imagery around target date"""
    
    start_date = (target_date - timedelta(days=days_buffer)).strftime('%Y-%m-%d')
    end_date = (target_date + timedelta(days=days_buffer)).strftime('%Y-%m-%d')
    
    print(f"Searching {platform.upper()} imagery from {start_date} to {end_date}")
    
    # Initialize STAC catalog
    catalog = pystac_client.Client.open(
        "https://planetarycomputer.microsoft.com/api/stac/v1/",
        modifier=pc.sign_inplace
    )
    
    # Use bounding box for search
    bounds = geometry.bounds
    bbox = [bounds[0], bounds[1], bounds[2], bounds[3]]
    
    # Search for Landsat data
    search = catalog.search(
        collections=["landsat-c2-l2"],
        bbox=bbox,
        datetime=f"{start_date}/{end_date}",
        query={
            "platform": {"in": [platform]},
            "eo:cloud_cover": {"lt": 50}
        }
    )
    
    items = list(search.items())
    
    if len(items) == 0:
        print("No images found with cloud filter, trying without...")
        search = catalog.search(
            collections=["landsat-c2-l2"],
            bbox=bbox,
            datetime=f"{start_date}/{end_date}",
            query={"platform": {"in": [platform]}}
        )
        items = list(search.items())
    
    if len(items) > 0:
        # Sort by cloud cover and date proximity
        items_with_info = []
        for item in items:
            cloud_cover = item.properties.get('eo:cloud_cover', 100)
            item_date = pd.to_datetime(item.properties['datetime'])
            
            if item_date.tz is not None:
                item_date = item_date.tz_localize(None)
            if hasattr(target_date, 'tz') and target_date.tz is not None:
                target_date = target_date.tz_localize(None)
            
            date_diff = abs((item_date - target_date).days)
            items_with_info.append((item, cloud_cover, date_diff, item_date.strftime('%Y-%m-%d')))
        
        items_with_info.sort(key=lambda x: (x[2], x[1]))
        sorted_items = [item for item, cloud, date_diff, date_str in items_with_info]
        
        print(f"✓ Found {len(sorted_items)} images")
        for i, (item, cloud, date_diff, date_str) in enumerate(items_with_info[:3]):
            print(f"  Image {i+1}: {date_str}, {cloud:.1f}% clouds, {date_diff} days from target")
        
        return sorted_items
    else:
        print("✗ No images found")
        return []

def search_with_expanding_dates(geometry, target_date, platform="landsat-5"):
    """Search with progressively expanding date ranges"""
    buffers = [30, 60, 120, 240, 365]
    
    for attempt, buffer in enumerate(buffers):
        print(f"\nAttempt {attempt + 1}: Searching with ±{buffer} day buffer...")
        items = search_landsat_imagery(geometry, target_date, days_buffer=buffer, platform=platform)
        
        if len(items) > 0:
            return items
            
        if attempt < len(buffers) - 1:
            print("No results, expanding search window...")
    
    print(f"❌ No {platform.upper()} imagery found")
    return []

def calculate_landsat_nbr(item, fire_polygon, satellite="landsat-5"):
    """Calculate NBR from Landsat data, clipped to fire boundary"""
    try:
        # Sign URLs with Planetary Computer
        signed_item = pc.sign(item)
        
        print(f"Processing {satellite.upper()} image from: {item.properties.get('datetime', 'Unknown')}")
        print(f"Cloud cover: {item.properties.get('eo:cloud_cover', 'Unknown')}%")
        
        # Define band mappings for different Landsat satellites
        band_mapping = {
            "landsat-5": {"nir": "nir08", "swir2": "swir22"},
            "landsat-7": {"nir": "nir08", "swir2": "swir22"},
            "landsat-8": {"nir": "nir08", "swir2": "swir22"},
            "landsat-9": {"nir": "nir08", "swir2": "swir22"}
        }
        
        bands = band_mapping.get(satellite, band_mapping["landsat-5"])
        nir_band = bands["nir"]
        swir2_band = bands["swir2"]
        
        # Check if bands exist
        available_assets = list(signed_item.assets.keys())
        print(f"Available assets: {available_assets}")
        
        if nir_band not in available_assets or swir2_band not in available_assets:
            print(f"❌ Required bands not found. Need: {nir_band}, {swir2_band}")
            return None
        
        print(f"Using NIR: {nir_band}, SWIR2: {swir2_band}")
        
        # Pre-crop to fire area for efficiency
        temp_gdf = gpd.GeoDataFrame([1], geometry=[fire_polygon], crs='EPSG:4326')
        
        # Load bands
        print(f"Loading {nir_band} band...")
        nir = rxr.open_rasterio(signed_item.assets[nir_band].href, chunks={'x': 1024, 'y': 1024})
        
        print(f"Loading {swir2_band} band...")
        swir2 = rxr.open_rasterio(signed_item.assets[swir2_band].href, chunks={'x': 1024, 'y': 1024})
        
        # Convert fire polygon to raster CRS and crop
        fire_proj = temp_gdf.to_crs(nir.rio.crs).geometry.iloc[0]
        bounds = fire_proj.bounds
        crop_buffer = 2000  # 2km buffer
        
        try:
            nir_cropped = nir.rio.clip_box(
                minx=bounds[0] - crop_buffer, miny=bounds[1] - crop_buffer,
                maxx=bounds[2] + crop_buffer, maxy=bounds[3] + crop_buffer
            )
            swir2_cropped = swir2.rio.clip_box(
                minx=bounds[0] - crop_buffer, miny=bounds[1] - crop_buffer,
                maxx=bounds[2] + crop_buffer, maxy=bounds[3] + crop_buffer
            )
            print(f"Cropped to fire area: NIR {nir_cropped.shape}, SWIR2 {swir2_cropped.shape}")
        except:
            print("Cropping failed, using original extent")
            nir_cropped = nir
            swir2_cropped = swir2
        
        # Landsat Collection 2 Level 2 scaling
        print("Applying Landsat Collection 2 scaling...")
        nir_scaled = nir_cropped * 0.0000275 - 0.2
        swir2_scaled = swir2_cropped * 0.0000275 - 0.2
        
        # Ensure same resolution
        if nir_scaled.shape != swir2_scaled.shape:
            print("Resampling SWIR2 to match NIR...")
            swir2_scaled = swir2_scaled.rio.reproject_match(nir_scaled)
        
        # Calculate NBR
        print("Calculating NBR...")
        nbr = (nir_scaled - swir2_scaled) / (nir_scaled + swir2_scaled)
        nbr = nbr.where((nir_scaled + swir2_scaled) != 0)
        
        # Clean up memory
        del nir, swir2, nir_cropped, swir2_cropped, nir_scaled, swir2_scaled
        import gc
        gc.collect()
        
        # Final clip to exact fire boundary
        print("Clipping to exact fire boundary...")
        nbr_fire_only = nbr.rio.clip([fire_proj], crs=nbr.rio.crs, drop=True, all_touched=True)
        
        del nbr
        gc.collect()
        
        # Check results
        valid_pixels = (~np.isnan(nbr_fire_only.values)).sum()
        total_pixels = nbr_fire_only.size
        
        print(f"Fire-only NBR shape: {nbr_fire_only.shape}")
        print(f"Valid pixels: {valid_pixels}/{total_pixels} ({100*valid_pixels/total_pixels:.1f}%)")
        
        if valid_pixels == 0:
            print("❌ No valid data within fire boundary")
            return None
        
        # Add metadata
        nbr_fire_only.attrs.update({
            'long_name': f'{satellite.upper()} NBR (Fire Boundary Only)',
            'description': '(NIR - SWIR2) / (NIR + SWIR2)',
            'satellite': satellite,
            'date': item.properties['datetime'],
            'valid_pixels': int(valid_pixels)
        })
        
        print(f"✓ NBR calculated successfully!")
        return nbr_fire_only
        
    except Exception as e:
        print(f"❌ Error calculating NBR: {e}")
        import traceback
        traceback.print_exc()
        return None

# =============================================================================
# ELEVATION DATA FUNCTIONS
# =============================================================================

def get_elevation_data(fire_geometry, buffer_meters=5000):
    """Get elevation data using OpenTopoData API"""
    print("Downloading elevation data...")
    
    # Get bounds
    fire_utm = fire_geometry.to_crs('EPSG:3857')
    buffered = fire_utm.buffer(buffer_meters)
    buffered_wgs84 = buffered.to_crs('EPSG:4326')
    bounds = buffered_wgs84.bounds
    if hasattr(bounds, 'iloc'):
        bounds = bounds.iloc[0]
    
    try:
        print("Using OpenTopoData API for elevation...")
        
        # Create grid of points for elevation sampling
        lat_points = np.linspace(bounds[1], bounds[3], 25)
        lon_points = np.linspace(bounds[0], bounds[2], 25)
        
        elevations = []
        locations = []
        
        # Sample elevation at grid points
        for i, lat in enumerate(lat_points):
            for j, lon in enumerate(lon_points):
                url = f"https://api.opentopodata.org/v1/srtm30m?locations={lat},{lon}"
                try:
                    response = requests.get(url, timeout=10)
                    if response.status_code == 200:
                        data = response.json()
                        if data['status'] == 'OK' and len(data['results']) > 0:
                            elev = data['results'][0]['elevation']
                            if elev is not None:
                                elevations.append(elev)
                                locations.append((lon, lat))
                except:
                    continue
                
                # Rate limiting
                import time
                time.sleep(0.05)
            
            # Progress indicator
            if i % 5 == 0:
                print(f"  Downloaded {len(elevations)} elevation points...")
        
        if len(elevations) > 20:
            print(f"✓ Retrieved {len(elevations)} elevation points")
            
            # Interpolate to create grid
            from scipy.interpolate import griddata
            
            # Create target grid
            target_lats = np.linspace(bounds[1], bounds[3], 150)
            target_lons = np.linspace(bounds[0], bounds[2], 150)
            target_lon_grid, target_lat_grid = np.meshgrid(target_lons, target_lats)
            
            # Interpolate elevations
            points = np.array(locations)
            values = np.array(elevations)
            
            interpolated = griddata(points, values, (target_lon_grid, target_lat_grid), method='cubic')
            
            # Fill any NaN values with nearest neighbor
            if np.isnan(interpolated).any():
                interpolated_nn = griddata(points, values, (target_lon_grid, target_lat_grid), method='nearest')
                interpolated = np.where(np.isnan(interpolated), interpolated_nn, interpolated)
            
            # Create xarray dataset
            elevation = xr.DataArray(
                interpolated,
                coords={'y': target_lats, 'x': target_lons},
                dims=['y', 'x']
            ).rio.write_crs('EPSG:4326')
            
            print(f"✓ Created elevation grid: {elevation.shape}")
            print(f"  Elevation range: {float(elevation.min().values):.0f}m to {float(elevation.max().values):.0f}m")
            return elevation
            
        else:
            print(f"❌ Only retrieved {len(elevations)} points, not enough for analysis")
            return None
            
    except Exception as e:
        print(f"❌ Error downloading elevation data: {e}")
        return None

# =============================================================================
# ELEVATION ANALYSIS FUNCTIONS
# =============================================================================

def analyze_fire_severity_elevation(dnbr, elevation, fire_geometry):
    """Analyze relationship between fire severity (dNBR) and elevation"""
    print("\nAnalyzing fire severity vs elevation relationship...")
    
    # Ensure both datasets have same CRS
    if dnbr.rio.crs != elevation.rio.crs:
        print("Reprojecting elevation to match dNBR...")
        elevation = elevation.rio.reproject_match(dnbr)
    
    # Clip both datasets to exact fire boundary
    fire_proj = gpd.GeoDataFrame([1], geometry=[fire_geometry], crs='EPSG:4326')
    fire_proj = fire_proj.to_crs(dnbr.rio.crs)
    
    print(f"Original shapes - dNBR: {dnbr.shape}, Elevation: {elevation.shape}")
    
    dnbr_fire = dnbr.rio.clip(fire_proj.geometry, drop=True, all_touched=True)
    elevation_fire = elevation.rio.clip(fire_proj.geometry, drop=True, all_touched=True)
    
    print(f"After clipping - dNBR: {dnbr_fire.shape}, Elevation: {elevation_fire.shape}")
    
    if dnbr_fire.shape != elevation_fire.shape:
        print("Reprojecting elevation to match dNBR grid...")
        elevation_fire = elevation_fire.rio.reproject_match(dnbr_fire)
    
    # Flatten and clean data
    dnbr_values = dnbr_fire.values.flatten()
    elev_values = elevation_fire.values.flatten()
    
    valid_mask = ~(np.isnan(dnbr_values) | np.isnan(elev_values))
    dnbr_clean = dnbr_values[valid_mask]
    elev_clean = elev_values[valid_mask]
    
    if len(dnbr_clean) == 0:
        print("❌ No overlapping valid data")
        return None
    
    print(f"✓ Analyzing {len(dnbr_clean)} valid pixels")
    
    # Calculate statistics
    correlation, p_value = stats.pearsonr(dnbr_clean, elev_clean)
    
    # Elevation binning analysis
    n_bins = 8
    elev_bins = np.linspace(elev_clean.min(), elev_clean.max(), n_bins + 1)
    elev_centers = (elev_bins[:-1] + elev_bins[1:]) / 2
    
    severity_by_elevation = []
    
    for i in range(n_bins):
        mask = (elev_clean >= elev_bins[i]) & (elev_clean < elev_bins[i + 1])
        if i == n_bins - 1:
            mask = (elev_clean >= elev_bins[i]) & (elev_clean <= elev_bins[i + 1])
        
        if np.sum(mask) > 0:
            bin_dnbr = dnbr_clean[mask]
            
            high_sev = np.sum(bin_dnbr > 0.66) / len(bin_dnbr) * 100
            mod_high_sev = np.sum((bin_dnbr >= 0.44) & (bin_dnbr <= 0.66)) / len(bin_dnbr) * 100
            mod_low_sev = np.sum((bin_dnbr >= 0.25) & (bin_dnbr < 0.44)) / len(bin_dnbr) * 100
            low_sev = np.sum((bin_dnbr >= 0.1) & (bin_dnbr < 0.25)) / len(bin_dnbr) * 100
            unburned = np.sum(bin_dnbr < 0.1) / len(bin_dnbr) * 100
            
            severity_by_elevation.append({
                'elevation': elev_centers[i],
                'high_severity': high_sev,
                'moderate_high': mod_high_sev,
                'moderate_low': mod_low_sev,
                'low_severity': low_sev,
                'unburned': unburned,
                'mean_dnbr': np.mean(bin_dnbr),
                'pixel_count': len(bin_dnbr)
            })
    
    results = {
        'dnbr_values': dnbr_clean,
        'elevation_values': elev_clean,
        'correlation': correlation,
        'p_value': p_value,
        'elevation_bins': elev_centers,
        'severity_by_elevation': pd.DataFrame(severity_by_elevation)
    }
    
    return results

def plot_elevation_analysis(results, fire_name="Fire"):
    """Create comprehensive plots of fire severity vs elevation"""
    fig = plt.figure(figsize=(16, 10))
    
    # 1. Scatter plot with trend line
    ax1 = plt.subplot(2, 3, 1)
    plt.scatter(results['elevation_values'], results['dnbr_values'], 
                alpha=0.5, s=2, c='red')
    
    z = np.polyfit(results['elevation_values'], results['dnbr_values'], 1)
    p = np.poly1d(z)
    x_trend = np.linspace(results['elevation_values'].min(), 
                         results['elevation_values'].max(), 100)
    plt.plot(x_trend, p(x_trend), 'blue', linewidth=2, alpha=0.8)
    
    plt.xlabel('Elevation (m)')
    plt.ylabel('dNBR')
    plt.title(f'Fire Severity vs Elevation\nr = {results["correlation"]:.3f}, p = {results["p_value"]:.4f}')
    plt.grid(True, alpha=0.3)
    
    # 2. Mean dNBR by elevation
    ax2 = plt.subplot(2, 3, 2)
    severity_df = results['severity_by_elevation']
    plt.plot(severity_df['elevation'], severity_df['mean_dnbr'], 
             'o-', linewidth=2, markersize=8, color='darkred')
    plt.xlabel('Elevation (m)')
    plt.ylabel('Mean dNBR')
    plt.title('Mean Fire Severity by Elevation')
    plt.grid(True, alpha=0.3)
    
    # 3. Stacked bar chart
    ax3 = plt.subplot(2, 3, 3)
    x_pos = np.arange(len(severity_df))
    
    plt.bar(x_pos, severity_df['high_severity'], label='High (>0.66)', color='darkred')
    plt.bar(x_pos, severity_df['moderate_high'], 
            bottom=severity_df['high_severity'], label='Mod-High (0.44-0.66)', color='red')
    plt.bar(x_pos, severity_df['moderate_low'], 
            bottom=severity_df['high_severity'] + severity_df['moderate_high'], 
            label='Mod-Low (0.25-0.44)', color='orange')
    plt.bar(x_pos, severity_df['low_severity'], 
            bottom=severity_df['high_severity'] + severity_df['moderate_high'] + severity_df['moderate_low'], 
            label='Low (0.1-0.25)', color='yellow')
    plt.bar(x_pos, severity_df['unburned'],
            bottom=severity_df['high_severity'] + severity_df['moderate_high'] + 
                   severity_df['moderate_low'] + severity_df['low_severity'],
            label='Unburned (<0.1)', color='green')
    
    plt.xlabel('Elevation Bin')
    plt.ylabel('Percentage')
    plt.title('Burn Severity by Elevation')
    plt.xticks(x_pos, [f"{elev:.0f}m" for elev in severity_df['elevation']], rotation=45)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
    
    # 4. Elevation histogram
    ax4 = plt.subplot(2, 3, 4)
    plt.hist(results['elevation_values'], bins=30, alpha=0.7, color='brown', edgecolor='black')
    plt.xlabel('Elevation (m)')
    plt.ylabel('Pixels')
    plt.title('Elevation Distribution')
    plt.grid(True, alpha=0.3)
    
    # 5. dNBR histogram
    ax5 = plt.subplot(2, 3, 5)
    plt.hist(results['dnbr_values'], bins=30, alpha=0.7, color='red', edgecolor='black')
    plt.xlabel('dNBR')
    plt.ylabel('Pixels')
    plt.title('dNBR Distribution')
    plt.grid(True, alpha=0.3)
    
    # 6. Summary statistics
    ax6 = plt.subplot(2, 3, 6)
    ax6.axis('off')
    
    corr_abs = abs(results['correlation'])
    strength = "Strong" if corr_abs > 0.7 else "Moderate" if corr_abs > 0.3 else "Weak"
    direction = "positive" if results['correlation'] > 0 else "negative"
    significance = "significant" if results['p_value'] < 0.05 else "not significant"
    
    summary_text = f"""ELEVATION ANALYSIS SUMMARY

Fire: {fire_name}
Pixels: {len(results['dnbr_values']):,}

Elevation:
• Range: {results['elevation_values'].min():.0f} - {results['elevation_values'].max():.0f} m
• Mean: {results['elevation_values'].mean():.0f} m

dNBR:
• Range: {results['dnbr_values'].min():.3f} - {results['dnbr_values'].max():.3f}
• Mean: {results['dnbr_values'].mean():.3f}

Correlation:
• r = {results['correlation']:.4f}
• p = {results['p_value']:.6f}
• {strength} {direction} correlation
• {significance} (α=0.05)"""
    
    ax6.text(0.05, 0.95, summary_text, transform=ax6.transAxes, fontsize=10,
             verticalalignment='top', fontfamily='monospace',
             bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.8))
    
    plt.suptitle(f'{fire_name} - Fire Severity vs Elevation Analysis', fontsize=16, y=0.98)
    plt.tight_layout()
    plt.subplots_adjust(top=0.93)
    plt.show()

# =============================================================================
# MAIN WORKFLOW FUNCTION
# =============================================================================

def analyze_real_fire_with_elevation(geojson_path, target_year=1989, fire_index=0, 
                                   satellite="landsat-5"):
    """
    Complete workflow: Load real fire, calculate dNBR, analyze with elevation
    """
    
    print("="*80)
    print("REAL FIRE SEVERITY + ELEVATION ANALYSIS")
    print("="*80)
    
    # 1. Load and select fire
    fire_info = load_and_select_fire(geojson_path, target_year, fire_index)
    
    # 2. Define analysis dates
    fire_year = fire_info['fire_year']
    pre_fire_date = pd.to_datetime(f"{fire_year}-04-15")
    post_fire_date = pd.to_datetime(f"{fire_year + 1}-07-01")
    
    print(f"\n{'='*60}")
    print("LANDSAT ANALYSIS TIMELINE")
    print(f"{'='*60}")
    print(f"Pre-fire (baseline): {pre_fire_date.strftime('%Y-%m-%d')}")
    print(f"Post-fire (1 year):  {post_fire_date.strftime('%Y-%m-%d')}")
    print(f"Satellite platform:  {satellite.upper()}")
    
    # 3. Search for pre-fire imagery
    print(f"\n{'='*60}")
    print("SEARCHING FOR PRE-FIRE IMAGERY")
    print(f"{'='*60}")
    pre_fire_items = search_with_expanding_dates(fire_info['geometry'], pre_fire_date, satellite)
    
    if len(pre_fire_items) == 0:
        print("❌ No pre-fire imagery found")
        return None
    
    # 4. Calculate pre-fire NBR
    print(f"\n{'='*60}")
    print("CALCULATING PRE-FIRE NBR")
    print(f"{'='*60}")
    pre_fire_nbr = calculate_landsat_nbr(pre_fire_items[0], fire_info['geometry'], satellite)
    
    if pre_fire_nbr is None:
        print("❌ Pre-fire NBR calculation failed")
        return None
    
    pre_stats = pre_fire_nbr.compute()
    print(f"✓ Pre-fire NBR: Mean {float(pre_stats.mean().values):.3f}, "
          f"Range {float(pre_stats.min().values):.3f} to {float(pre_stats.max().values):.3f}")
    
    # 5. Search for post-fire imagery
    print(f"\n{'='*60}")
    print("SEARCHING FOR POST-FIRE IMAGERY")
    print(f"{'='*60}")
    post_fire_items = search_with_expanding_dates(fire_info['geometry'], post_fire_date, satellite)
    
    if len(post_fire_items) == 0:
        print("❌ No post-fire imagery found")
        return None
    
    # 6. Calculate post-fire NBR
    print(f"\n{'='*60}")
    print("CALCULATING POST-FIRE NBR")
    print(f"{'='*60}")
    post_fire_nbr = calculate_landsat_nbr(post_fire_items[0], fire_info['geometry'], satellite)
    
    if post_fire_nbr is None:
        print("❌ Post-fire NBR calculation failed")
        return None
    
    post_stats = post_fire_nbr.compute()
    print(f"✓ Post-fire NBR: Mean {float(post_stats.mean().values):.3f}, "
          f"Range {float(post_stats.min().values):.3f} to {float(post_stats.max().values):.3f}")
    
    # 7. Calculate dNBR
    print(f"\n{'='*60}")
    print("CALCULATING dNBR")
    print(f"{'='*60}")
    
    # Ensure matching spatial extent
    if pre_fire_nbr.rio.crs != post_fire_nbr.rio.crs:
        post_fire_nbr = post_fire_nbr.rio.reproject_match(pre_fire_nbr)
    
    dnbr = pre_fire_nbr - post_fire_nbr
    dnbr.attrs.update({
        'long_name': f'{satellite.upper()} dNBR (Fire Boundary Only)',
        'description': 'Pre-fire NBR - Post-fire NBR',
        'fire_id': fire_info['fire_id'],
        'incident_name': fire_info['incident_name'],
        'fire_year': fire_year,
        'satellite': satellite
    })
    
    # Calculate burn severity statistics
    dnbr_computed = dnbr.compute()
    dnbr_flat = dnbr_computed.values.flatten()
    dnbr_flat = dnbr_flat[~np.isnan(dnbr_flat)]
    
    if len(dnbr_flat) > 0:
        # Calculate severity classes
        high_severity = np.sum(dnbr_flat > 0.66)
        mod_high_severity = np.sum((dnbr_flat >= 0.44) & (dnbr_flat <= 0.66))
        mod_low_severity = np.sum((dnbr_flat >= 0.25) & (dnbr_flat < 0.44))
        low_severity = np.sum((dnbr_flat >= 0.1) & (dnbr_flat < 0.25))
        unburned = np.sum(dnbr_flat < 0.1)
        total_pixels = len(dnbr_flat)
        
        print(f"✓ dNBR calculated successfully!")
        print(f"  Mean dNBR: {np.mean(dnbr_flat):.3f}")
        print(f"  Range: {np.min(dnbr_flat):.3f} to {np.max(dnbr_flat):.3f}")
        print(f"  Total pixels: {total_pixels:,}")
        
        print(f"\nBurn Severity Classification:")
        print(f"  High Severity (>0.66):      {high_severity:5d} pixels ({100*high_severity/total_pixels:5.1f}%)")
        print(f"  Moderate-High (0.44-0.66):  {mod_high_severity:5d} pixels ({100*mod_high_severity/total_pixels:5.1f}%)")
        print(f"  Moderate-Low (0.25-0.44):   {mod_low_severity:5d} pixels ({100*mod_low_severity/total_pixels:5.1f}%)")
        print(f"  Low Severity (0.1-0.25):    {low_severity:5d} pixels ({100*low_severity/total_pixels:5.1f}%)")
        print(f"  Unburned (<0.1):            {unburned:5d} pixels ({100*unburned/total_pixels:5.1f}%)")
    
    # 8. Get elevation data
    print(f"\n{'='*60}")
    print("DOWNLOADING ELEVATION DATA")
    print(f"{'='*60}")
    fire_gdf = gpd.GeoDataFrame([1], geometry=[fire_info['geometry']], crs='EPSG:4326')
    elevation = get_elevation_data(fire_gdf, buffer_meters=3000)
    
    if elevation is None:
        print("❌ Could not download elevation data")
        return None
    
    # 9. Perform elevation analysis
    print(f"\n{'='*60}")
    print("ANALYZING FIRE SEVERITY vs ELEVATION")
    print(f"{'='*60}")
    elevation_results = analyze_fire_severity_elevation(dnbr, elevation, fire_info['geometry'])
    
    if elevation_results is None:
        print("❌ Elevation analysis failed")
        return None
    
    # 10. Create visualizations
    print(f"\n{'='*60}")
    print("CREATING VISUALIZATIONS")
    print(f"{'='*60}")
    fire_display_name = f"{fire_info['incident_name']} ({fire_year})"
    plot_elevation_analysis(elevation_results, fire_display_name)
    
    # 11. Print detailed results
    print(f"\n{'='*80}")
    print("FINAL ANALYSIS RESULTS")
    print(f"{'='*80}")
    
    print(f"\nFire Information:")
    print(f"  Name: {fire_info['incident_name']}")
    print(f"  ID: {fire_info['fire_id']}")
    print(f"  Year: {fire_year}")
    print(f"  Satellite: {satellite.upper()}")
    
    print(f"\nSpatial Analysis:")
    print(f"  Total analyzed pixels: {len(elevation_results['dnbr_values']):,}")
    print(f"  Fire area: {fire_info['geometry'].area:.8f} square degrees")
    
    print(f"\nElevation Summary:")
    print(f"  Range: {elevation_results['elevation_values'].min():.0f} - {elevation_results['elevation_values'].max():.0f} m")
    print(f"  Mean: {elevation_results['elevation_values'].mean():.0f} m")
    print(f"  Elevation Relief: {elevation_results['elevation_values'].max() - elevation_results['elevation_values'].min():.0f} m")
    
    print(f"\nFire Severity Summary:")
    print(f"  dNBR Range: {elevation_results['dnbr_values'].min():.3f} - {elevation_results['dnbr_values'].max():.3f}")
    print(f"  Mean dNBR: {elevation_results['dnbr_values'].mean():.3f}")
    
    print(f"\nElevation-Fire Severity Correlation:")
    print(f"  Pearson correlation: {elevation_results['correlation']:.4f}")
    print(f"  P-value: {elevation_results['p_value']:.6f}")
    print(f"  Statistical significance: {'Yes' if elevation_results['p_value'] < 0.05 else 'No'} (α=0.05)")
    
    # Interpretation
    corr_abs = abs(elevation_results['correlation'])
    if corr_abs > 0.3:
        strength = "strong" if corr_abs > 0.7 else "moderate"
    else:
        strength = "weak"
    
    direction = "positive" if elevation_results['correlation'] > 0 else "negative"
    
    print(f"\nInterpretation:")
    print(f"  There is a {strength} {direction} correlation between elevation and fire severity.")
    
    if elevation_results['correlation'] > 0.1:
        print(f"  Higher elevations experienced more severe burning on average.")
        print(f"  This could indicate:")
        print(f"    - Drier conditions at higher elevations")
        print(f"    - Different fuel types/loading with elevation")
        print(f"    - Topographic wind effects enhancing fire behavior")
    elif elevation_results['correlation'] < -0.1:
        print(f"  Lower elevations experienced more severe burning on average.")
        print(f"  This could indicate:")
        print(f"    - Denser/drier fuels at lower elevations")
        print(f"    - Fire spread patterns influenced by topography")
        print(f"    - Climate/moisture gradients with elevation")
    else:
        print(f"  Elevation does not strongly influence fire severity in this fire.")
        print(f"  Other factors (weather, fuels, ignition patterns) may be more important.")
    
    # Elevation-based severity table
    severity_df = elevation_results['severity_by_elevation']
    print(f"\nDetailed Severity by Elevation Zones:")
    print(f"{'Elevation (m)':<12} {'Mean dNBR':<10} {'High Sev %':<11} {'Mod-High %':<11} {'Pixels':<8}")
    print("-" * 65)
    for _, row in severity_df.iterrows():
        print(f"{row['elevation']:<12.0f} {row['mean_dnbr']:<10.3f} "
              f"{row['high_severity']:<11.1f} {row['moderate_high']:<11.1f} {row['pixel_count']:<8.0f}")
    
    print(f"\n✓ ANALYSIS COMPLETE!")
    print(f"{'='*80}")
    
    # Return all results for further analysis if needed
    return {
        'fire_info': fire_info,
        'pre_fire_nbr': pre_fire_nbr,
        'post_fire_nbr': post_fire_nbr,
        'dnbr': dnbr,
        'elevation': elevation,
        'elevation_results': elevation_results,
        'satellite': satellite
    }

# =============================================================================
# EXAMPLE USAGE
# =============================================================================

def run_analysis_example():
    """
    Example of how to run the complete analysis
    """
    
    # Configuration
    geojson_path = "../downloaded_data/mnf_fires_all.geojson"
    target_year = 1989
    fire_index = 0  # Select which fire from the year (0 = first fire)
    satellite = "landsat-5"  # For 1989 fires
    
    print("Starting real fire severity + elevation analysis...")
    print(f"Target year: {target_year}")
    print(f"Fire index: {fire_index}")
    print(f"Satellite: {satellite}")
    
    # Run the complete analysis
    results = analyze_real_fire_with_elevation(
        geojson_path=geojson_path,
        target_year=target_year,
        fire_index=fire_index,
        satellite=satellite
    )
    
    if results is not None:
        print("\n" + "="*80)
        print("SUCCESS! Analysis completed.")
        print("="*80)
        print("\nResults include:")
        print("- Real fire geometry and metadata")
        print("- Landsat-derived pre/post-fire NBR")
        print("- dNBR (burn severity) analysis")
        print("- Real SRTM elevation data")
        print("- Statistical correlation analysis")
        print("- Comprehensive visualizations")
        
        return results
    else:
        print("\n" + "="*80)
        print("ANALYSIS FAILED - Check error messages above")
        print("="*80)
        return None

# =============================================================================
# ALTERNATIVE FIRE SELECTION HELPER
# =============================================================================

def explore_available_fires(geojson_path, target_year=1989):
    """
    Helper function to explore available fires before running analysis
    """
    print("="*60)
    print("EXPLORING AVAILABLE FIRES")
    print("="*60)
    
    # Load fire data
    fires = gpd.read_file(geojson_path)
    
    # Clean fire years
    def clean_fire_year(year_value):
        try:
            if pd.isna(year_value):
                return None
            year_str = str(year_value)
            if '-' in year_str:
                return int(year_str.split('-')[0])
            return int(float(year_str))
        except (ValueError, TypeError):
            return None
    
    fires['FIRE_YEAR_CLEAN'] = fires['FIRE_YEAR'].apply(clean_fire_year)
    fires_clean = fires.dropna(subset=['FIRE_YEAR_CLEAN'])
    fires_clean['FIRE_YEAR_CLEAN'] = fires_clean['FIRE_YEAR_CLEAN'].astype(int)
    
    # Show available years
    available_years = sorted(fires_clean['FIRE_YEAR_CLEAN'].unique())
    print(f"Available fire years: {available_years}")
    
    # Show fires for target year and nearby years
    for year in range(target_year-2, target_year+3):
        year_fires = fires_clean[fires_clean['FIRE_YEAR_CLEAN'] == year]
        if len(year_fires) > 0:
            print(f"\n{year} fires ({len(year_fires)} total):")
            year_fires_wgs84 = year_fires.to_crs('EPSG:4326')
            
            for i, (idx, fire) in enumerate(year_fires_wgs84.iterrows()):
                fire_name = fire.get('INCIDENT', f'Fire_{i}')
                fire_id = fire.get('UNQE_FIRE_', f'ID_{i}')
                area = fire.geometry.area
                print(f"  Index {i}: {fire_name} ({fire_id}) - Area: {area:.8f} sq deg")
    
    print(f"\nTo analyze a specific fire, use:")
    print(f"analyze_real_fire_with_elevation('{geojson_path}', target_year=YEAR, fire_index=INDEX)")

if __name__ == "__main__":
    print("Real Fire Severity + Elevation Analysis Workflow")
    print("="*50)
    print("Main functions:")
    print("1. explore_available_fires() - See what fires are available")
    print("2. analyze_real_fire_with_elevation() - Run complete analysis")
    print("3. run_analysis_example() - Run with default settings")
    print("\nExample usage:")
    print("results = run_analysis_example()")
```
